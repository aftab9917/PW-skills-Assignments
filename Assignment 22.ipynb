```python
#ANSWER 2

'''A scatter plot is a type of plot that displays values for two variables as points on a two-dimensional plane. The position of each point is determined by the value of the two variables. Here's the code to generate and plot a scatter plot:'''


import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#ANSWER 3

'''The subplot() function in Matplotlib is used to create a grid of subplots in a single figure. It allows you to plot multiple plots in one figure. Here's how you can draw four line plots using the subplot() function:'''


import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```

```python
#ANSWER 4

''' A bar plot is a graphical display of data using bars of different heights or lengths. It is used to compare different categories of data or to visualize the distribution of a dataset. Here's how you can plot a bar plot and a horizontal bar plot using the given data:'''


import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(company, profit)
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')

plt.subplot(1, 2, 2)
plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.tight_layout()
plt.show()
```

```python
#ANSWER 5

''' A box plot, also known as a box-and-whisker plot, is a graphical method for summarizing the distribution of a dataset. It displays the distribution of a dataset based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It is used to identify outliers and to compare the distribution of different datasets. Here's how you can plot a box plot using the given data:'''


import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Values')

plt.show()
```
